```python
connections = {
    "|" : [(0, 1), (0, -1)],
    "-" : [(1, 0), (-1, 0)],
    "L" : [(0, -1), (1, 0)],
    "J" : [(0, -1), (-1, 0)],
    "7" : [(0, 1), (-1, 0)],
    "F" : [(0, 1), (1, 0)],
    "." : []
}

test_1 = """.....
.S-7.
.|.|.
.L-J.
....."""

test_2 = """..F7.
.FJ|.
SJ.L7
|F--J
LJ..."""

test_3 = """
...........
.S-------7.
.|F-----7|.
.||.....||.
.||.....||.
.|L-7.F-J|.
.|..|.|..|.
.L--J.L--J.
..........."""

def add(a, b):
    return tuple(sum(x) for x in zip(a,b))

def parse_in(pipes_str=None):
    pipes = {}
    if not pipes_str:
        pipes_str = open("Inputs/Day10.txt").read()
    lines = pipes_str.strip().split("\n")
    for y, line in enumerate(lines):
        for x, tile in enumerate(line):
            if tile == "S":
                cons = []
                if lines[y-1][x] in "|7F":
                    cons.append((0, -1))
                if lines[y+1][x] in "|LJ":
                    cons.append((0, 1))
                if lines[y][x+1] in "-J7":
                    cons.append((1, 0))
                if lines[y][x-1] in "-LF":
                    cons.append((-1, 0))
                start = (x, y)
                pipes[(x, y)] = [add((x, y), i) for i in cons]
                #find start's shape
            else:
                pipes[(x, y)] = [add((x, y), i) for i in connections[tile]]
    return start, pipes

def parse_in2(pipes_str=None):
    if not pipes_str:
        pipes_str = open("Inputs/Day10.txt").read()
    lines = pipes_str.strip().split("\n")
    
    pipes = [[x for x in l] for l in lines]
    for y, line in enumerate(lines):
        for x, tile in enumerate(line):
            if tile == "S":
                cons = []
                if lines[y-1][x] in "|7F":
                    cons.append((0, -1))
                if lines[y+1][x] in "|LJ":
                    cons.append((0, 1))
                if lines[y][x+1] in "-J7":
                    cons.append((1, 0))
                if lines[y][x-1] in "-LF":
                    cons.append((-1, 0))
                
                for k,v in connections.items():
                    if all(c in v for c in cons):
                        #print("S is", k)
                        pipes[y][x] = k
                        
    
    
    return pipes

parse_in()
```
